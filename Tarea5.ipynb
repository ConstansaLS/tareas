Un cierto sensor se deja abierto por un segundo. El sensor consiste en $r$ pixeles y en promedio el sensor es golpeado por $2$ partículas por segundo, tal que la probabilidad de detectar una partícula en un pixel dado es de $2/r$ (para $r \gg 2$). Grafique la distribución de probabilidad de observar $n$ partículas para un pixel dado, en función de $n$, para diferentes valores de $r$:

* Usando la distribución binomial.

* ¿Cuándo es posible usar la distribución de Poisson?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python

def binomial(n, r, p):
  '''calcula la probabilidad de  obtener  "n"  resultados positivos "p" cuando se realiza el experimento  "r"  veces de manera independiente.'''
  return comb(r, n)*p**n*(1-p)**(r - n)
```

```python
def probs(r):
    '''hace una lista para la probabilidad de obesrvar n partículas para cada valor de r, usando la distribución binomial.
    Para este ejercicio, p=2/r'''
    return [binomial(n, r, 2/r) for n in range(r+1)]
```

```python
#Gráficos de la distribución binomial de probabilidades para diferentes valores de r
fig = plt.figure(figsize=(8,20), dpi= 100)

ax1 = fig.add_subplot(8,2,1)
ax2 = fig.add_subplot(8,2,2)
ax3 = fig.add_subplot(8,2,3)
ax4 = fig.add_subplot(8,2,4)
ax5 = fig.add_subplot(8,2,5)
ax6 = fig.add_subplot(8,2,6)
ax7 = fig.add_subplot(8,2,7)
ax8 = fig.add_subplot(8,2,8)

#eje x = n, número de partículas
#eje y = probabilidad de que n partículas golpeen un pixel
ax1.bar(range(4+1), probs(4))
ax1.set_xlabel("n (número de partículas)")
ax1.set_ylabel("probabilidad")
ax2.bar(range(5+1), probs(5))
ax2.set_xlabel("n (número de partículas)")
ax2.set_ylabel("probabilidad")
ax3.bar(range(10+1), probs(10))
ax3.set_xlabel("n (número de partículas)")
ax3.set_ylabel("probabilidad")
ax4.bar(range(15+1), probs(15))
ax4.set_xlabel("n (número de partículas)")
ax4.set_ylabel("probabilidad")
ax5.bar(range(20+1), probs(20))
ax5.set_xlabel("n (número de partículas)")
ax5.set_ylabel("probabilidad")
ax6.bar(range(25+1), probs(25))
ax6.set_xlabel("n (número de partículas)")
ax6.set_ylabel("probabilidad")
ax7.bar(range(30+1), probs(30))
ax7.set_xlabel("n (número de partículas)")
ax7.set_ylabel("probabilidad")
ax8.bar(range(50+1), probs(50))
ax8.set_xlabel("n (número de partículas)")
ax8.set_ylabel("probabilidad")

ax1.set_title('r = 4')
ax2.set_title('r = 5')
ax3.set_title('r = 10')
ax4.set_title('r = 15')
ax5.set_title('r = 20')
ax6.set_title('r = 25')
ax7.set_title('r = 30')
ax8.set_title('r = 50')

fig.tight_layout(pad=2.0)


fig.show()
```
